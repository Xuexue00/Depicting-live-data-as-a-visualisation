```python
import pandas as pd
from datetime import datetime, timedelta
from sdv.metadata import SingleTableMetadata
from sdv.lite import SingleTablePreset

# 加载原始数据
original_data_path = "emissions-1.csv"
original_data = pd.read_csv(original_data_path)

# 检测元数据
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(original_data)

# 使用预设生成器
synthesizer = SingleTablePreset(metadata, name='FAST_ML')

# 拟合数据
synthesizer.fit(data=original_data)

# 生成合成数据
synthetic_data = synthesizer.sample(num_rows=500)

# 定义生成时间戳的函数
def generate_timestamps(start_time, num_rows):
    return [start_time + timedelta(days=i) for i in range(num_rows)]

# 获取原始数据的起始时间
start_time = datetime.strptime("2020-03-23T09:26:11", "%Y-%m-%dT%H:%M:%S")

# 生成新的时间戳列并赋值给合成数据
synthetic_data['timestamp'] = generate_timestamps(start_time, len(synthetic_data))

# 显示更新后的合成数据
print(synthetic_data.head())

# 保存更新后的合成数据
synthetic_data.to_csv('synthetic_data.csv', index=False)
```
